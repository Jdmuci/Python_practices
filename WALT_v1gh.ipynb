### WALT - Cálculo simple

- Calculado en años respecto a la renta anual pagada por tenant.

File "WALT_v1.csv"

```python
import pandas as pd
import datetime as dt
```

```python
#Cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/Jdmuci/Python_practices/refs/heads/main/WALT_v1.csv")
print(df)

#df = pd.read_excel("WALT v1.xlsx")
```

```python
## Obtener Años Restantes y Renta Anual

#### INPUT Commencement Date ####
df["Fecha Inicial Considerada"] = pd.to_datetime("2024-12-01") ##### Agregar FECHA INICIAL
####

df["Fin Contrato"] = pd.to_datetime(df["Fin Contrato"]) # Dar formato de fecha a columna "Fin Contrato"

df["Dias Restantes"] = (df["Fin Contrato"] - df["Fecha Inicial Considerada"]).dt.days   # Calcular dias restantes
df["Años Restantes"] = (df["Dias Restantes"] / 365).round(3)


## Obtener Renta Anual
df["Renta Anual"] = (df["Renta mxn por sqm"] * df["GLA (sqm)"] * 12) #.apply(lambda x: f"{x:,.2f}")  #Separar con "," por cada mil

df.head(5)
```

```python
# Agregar Columna de WALT

df["Renta Anual"] = pd.to_numeric(df["Renta Anual"]) #Formato numerico a Renta Anual

df["WALT (años)"] = (df["Años Restantes"] * df["Renta Anual"]) / (df["Renta Anual"].sum())

# Se puede poner la suma de la renta anual en otra variable: Renta_Anual_Total = df['Renta Anual'].sum()

df.head(5)
```

```python
## Resultado:

WALT = df["WALT (años)"].sum().round(2)

print('-' * 21)
print(f'{"WALT:":<10} {WALT:<5}años')
print('-' * 21)
```
