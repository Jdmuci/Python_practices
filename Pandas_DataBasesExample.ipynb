Cargar datos [Load data]

```python
#Cargar datos

df = pd.read_csv("https://raw.githubusercontent.com/Jdmuci/Python_practices/main/pkmn_data.csv")
df.head(5)

#Si el archivo es de texto "txt" se pone el delimitador que tenga, en este caso "\t"
#df = pd.read_csv("pokemon_data.txt", delimiter = "\t")

#df_xlsx = pd.read_excel("pokemon_data.xlsx")
#df_xlsx.head(5)
```

Obtener columnas, filas y celdas específicas [Get specific columns, rows or cells from a Data Frame]

```python
#Leer encabezados de cada columna:
print(df.columns)

#Leer/onbtener columnas específicas:
print(df["Name"][0:3])  # 1 columna
print(df[["Name","Attack","Defense"]][0:3])  # 2 o más

#Leer/onbtener cada row:
print(df.iloc[1:3])  #Renglones del 1 al 3

#for index, row in df.iterrows():
#    print(index, row["Name"])

#Leer/ontener el contenido de celda específico (R,C) empiezan de 0
print(df.iloc[4,1])

```

```python
#Obtener data específica "filtrada" de una columna
#En este caso Fire es uno de los tipos dentro Type 1
df.loc[df["Type 1"] == "Fire"]
```

Describe and sort Functions

```python
#Describe
df.describe()
```

```python
#sort(ordenar de mayor a menor y viceversa)
df.sort_values(["Attack", "HP"], ascending = [1,0])   #por defecto es de - a +,   para cambiar es con     ascending = False
```

Make changes to data frame (Add/remove columns)

```python
# Agregar columna

#df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]

df["Total"] = df.iloc[:, 4:10].sum(axis=1)

df.head(5)

# remover la columna
#df = df.drop(columns=["Total"])


#Reordenar columnas:
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

Save your file in a new csv

```python
#Guardar en un nuevo archivo
df.to_csv("pokemon_data_modificado.csv", index = False)
#df.to_excel("pokemon_data_modificado.csv", index = False)
#df.to_excel("pokemon_data_modificado.csv", index = False, sep = "\t")
```

Filtering Data

```python
#Obtener data específica "filtrada" de una columna

#En este caso Fire es uno de los tipos dentro Type 1
#df.loc[df["Type 1"] == "Grass"]

new_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
new_df

#Para resetear index
new_df.reset_index(drop= True, inplace= True)
new_df
```

```python
#Filtrar por todos los nombres que contengan cierta palabra
df.loc[df["Name"].str.contains("Mega")]
```

Hacer un DF con Name, HP, Attack, Defense y Total con una columna que divida en 3 clases por HP; < 70 plata, de 71 a 100 son oro y > 100 son legend

```python
# 1 - DF
df1 = df[["Name","HP","Attack","Defense","Total"]]
df1
```

```python
# 2 - Segmentación por clase y agregar columna
df1["Clase"] = df1["HP"].apply(lambda x: "Plata" if x < 70 else("Oro" if 70 < x < 100 else "Legend"))
df1
```

```python
# 3 - Ordenar conforme a HP
df1.sort_values(["HP"])
```
